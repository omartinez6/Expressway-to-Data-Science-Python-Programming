```python
markers = ['o','*','.',',','x','X','+','p','P','s','d','D','H','h','v','^','<','>','1','2','3','4','|','_']

for x in markers:
    print(x)
    plt.plot(y,x)
    plt.show()
```

```python
y = np.array([1,3,13])
line = ['-',':','--','-.']
for x in line:
    print(x)
    x = 'o' + x + 'r'
    plt.plot(y,x)
    plt.show()
```

```python
color = ['r','g','b','c','m','y','k','w']
for x in color:
    print(x)
    x = 'o:' + x
    plt.plot(y,x)
    plt.show()
```

```python
x = np.arange(10)
y1 = x * 2
y2 = x ** 2
plt.plot(y1)
plt.plot(y2)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
plt.plot(x, y1, label = 'double')
plt.plot(x, y2, label = 'square')
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.grid()
plt.legend(['1st','2nd'])
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y1, label = 'double')
plt.legend()
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.subplot(2, 1, 2)
plt.plot(x, y2, label = 'square')
plt.legend()
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.grid()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y1, label = 'double')
plt.legend()
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.subplot(1, 2, 2)
plt.plot(x, y2, label = 'square')
plt.legend()
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.grid()
```

```python
x = np.arange(10)
alpha = (x + 1) ** 2 / 2
delta = (x + 2) ** 2
omicron = (x + 3) ** 2 * 3
plt.plot(x, alpha, label = 'Alpha')
plt.plot(x, delta, label = 'Delta')
plt.plot(x, omicron, label = 'Omicron')
plt.legend()
plt.grid()
plt.xlabel('Xs')
plt.ylabel('YS')
```
