## Set up the environment

### Task: import packages and rename them accordingly

````python
# your code is here
import seaborn as sns
import pandas as pd
import numpy as np
````

## Load the dataset

````python
# your code is here
iris = sns.load_dataset('iris')
iris.head()
````

## Play with relational plots

````python
# your code is here
sns.relplot(x='sepal_length', y='sepal_width', data=iris, hue='species', kind='scatter')
````

````python
sns.relplot(x=range(len(iris)), y='petal_length', data=iris, hue='species', kind='line')
````

````python
sns.relplot(x='sepal_length', y='sepal_width', data=iris, col='species')
````

````python
sns.relplot(x='petal_length', y='petal_width', data=iris, size='sepal_length', hue='species')
````

## Play with distribution plots

````python
# your code is here
sns.displot(data=iris, x='sepal_length', kde=False)
````

````python
sns.displot(data=iris, x='sepal_length', kde=True, rug=True)
````

````python
sns.displot(data=iris, x='sepal_length', y='sepal_width', kind='kde')
````

## Play with other plots

````python
# your code is here
sns.boxplot(data=iris, x='species', y='sepal_length')
````

````python
sns.violinplot(data=iris, x='species', y='sepal_length')
````

````python
sns.swarmplot(data=iris, x='species', y='sepal_length')
````

## Additional Challenge
You may want to create new columns based on existing data using Pandas. You can play with the new columns as well.

````python
# your code is here
corr = np.corrcoef(iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']], rowvar=False)
sns.heatmap(corr, annot=True, cmap='coolwarm')
````

````python
sns.pairplot(data=iris, hue='species')
````
