### Task: import packages and rename them accordingly

````python
# your code is here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

## Load the dataset

````python
# your code is here
data = sns.load_dataset('diamonds')
data.head()
````

````python
carat = data['carat']
price = data['price']
````

## Play with distribution plots
You can show the data distribution for some attributes interest you.

````python
# your code is here
plt.subplot(1, 2, 1)
plt.boxplot(price)
plt.ylabel('price')
plt.subplot(1, 2, 2)
plt.hist(carat,bins=20)
plt.xlabel('carat')
````

## Play with relational plots
You can show how two or more attributes are associated.

````python
# your code is here
plt.scatter(carat,price)
plt.xlabel('carat')
plt.ylabel('price')
plt.grid()
````

## Play with other plots
You can play with other plots we learned here. Be as creative as you can and see how the visualization helps you to understand the data better.

````python
# your code is here
plt.boxplot([data.loc[data['cut'] == 'Fair', 'price'],
             data.loc[data['cut'] == 'Good', 'price'],
             data.loc[data['cut'] == 'Very Good', 'price'],
             data.loc[data['cut'] == 'Premium', 'price'],
             data.loc[data['cut'] == 'Ideal', 'price']],
            labels=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
````

````python
data['clarity'].value_counts().plot(kind='bar')
plt.xlabel('Clarity')
plt.ylabel('Count')
plt.show()
````

````python
price_by_carat = data.groupby('carat')['price'].mean().reset_index()
plt.plot(price_by_carat['carat'], price_by_carat['price'])
plt.xlabel('Carat')
plt.ylabel('Mean price')
plt.show()
````

````python
data['cut'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.show()
````

````python
data_cut_clarity = data.groupby(['cut', 'clarity'])['price'].count().unstack()
data_cut_clarity.plot(kind='bar', stacked=True)
plt.xlabel('Cut')
plt.ylabel('Count')
plt.show()
````
